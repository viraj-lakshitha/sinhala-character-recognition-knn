### Sinhala Character Recognition Graphical User Interface

* Using Existing Model 'sinhala-character-knn.sav'

* Some of them are not-well recognized, because insufficient data

```python
import tkinter as tk
from PIL import ImageTk,Image,ImageDraw
import numpy as np # Import Numpy Array
import cv2 # Importing Open-CV
import joblib # Importing joblib library to re-use trained model

win=tk.Tk()

# Main parameters to create main window
w,h=500,500
fontButton='Helvetica 20 bold'
fontLabel='Helvetica 22 italic'
count=0
predictResult='NONE'

# Load the Model
model=joblib.load('sinhala-character-knn.sav')

# Create Funtion
def eventFuntion(event) :
    #print('EVENT')
    
    x=event.x # Getting X Cordinates
    y=event.y # Getting Y Cordinates
    
    x1=x-10
    y1=y-10
    
    x2=x+10
    y2=y+10
    
    canvas.create_oval((x1,y1,x2,y2),fill='black')
    imgDraw.ellipse((x1,y1,x2,y2),fill='black') # Background Color - Black

# Save Button Funtions
def save() :
    
    # Global Variable
    global count
    
    imgArray=np.array(img)
    imgArray=cv2.resize(imgArray,(8,8))
    
    cv2.imwrite('data/'+str(count)+'.jpg',imgArray)
    # Filename to Save data/0.jpg
    count=count+1
    
# Clear Button Funtions
def clear() :
    
    # Global Variable
    global img,imgDraw
    
    canvas.delete('all')
    img=Image.new('RGB',(w,h),(255,255,255))
    imgDraw=ImageDraw.Draw(img)
    

# Prdict Button Funtion
def predict() :
    
    global predictResult
    
    # Pre-process the image
    imgArray=np.array(img)
    # RGB to GRAY
    imgArray=cv2.cvtColor(imgArray,cv2.COLOR_RGB2GRAY)
    imgArray=cv2.resize(imgArray,(8,8))
    #Flatten Image
    imgArray=np.reshape(imgArray,(1,64))
    #Scaling 0-16
    imgArray=(imgArray/255)*16
    
    result=model.predict(imgArray)
    
    if (result==0) :
        predictResult='අ'
        
    elif (result==1) :
        predictResult='ඉ'
        
    elif (result==2) :
        predictResult='ඒ'
        
    elif (result==3) :
        predictResult='උ'
        
    labelStatus.config(text='PREDICTED LETTER : '+str(predictResult))

canvas=tk.Canvas(win,height=h,width=w,bg='white')
canvas.grid(row=0,column=0,columnspan=4) # pack,place and grid - Methods to locate the Canva in Main Window

# Create Buttons
buttonSave=tk.Button(win,text='SAVE',bg='light sea green',fg='white',font=fontButton,command=save)
buttonSave.grid(row=1,column=0)

buttonPredict=tk.Button(win,text='PREDICT',bg='blue',fg='white',font=fontButton,command=predict)
buttonPredict.grid(row=1,column=1)

buttonReset=tk.Button(win,text='RESET',bg='green',fg='white',font=fontButton,command=clear)
buttonReset.grid(row=1,column=2)

buttonExit=tk.Button(win,text='EXIT',bg='red',fg='white',font=fontButton,command=win.destroy)
buttonExit.grid(row=1,column=3)

#Create Label
labelStatus=tk.Label(win,text='PREDICTED LETTER : NONE',bg='white',fg='black',font=fontLabel)
labelStatus.grid(row=2,column=0,columnspan=4)


#B1-Motion = Left Click Button and B2-Motion =Scroll Button B3-Motion = RIght Button
canvas.bind('<B1-Motion>',eventFuntion)

#Create Image Object
img=Image.new('RGB',(w,h),(255,255,255))
imgDraw=ImageDraw.Draw(img)

# Create Main Window
win.mainloop()
```
